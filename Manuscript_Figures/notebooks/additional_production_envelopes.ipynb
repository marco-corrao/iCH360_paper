# Additional production envelopes

An extended set of production envelopes comparisons between iCH360 and its genome-scale parent (iML1515), focusing on additional key metabolic precursors not directly shown in the manuscript.

```python
import sys
sys.path.append("../../utils/")
import fba_utils

import cobra
import matplotlib.pyplot as plt
import importlib
importlib.reload(fba_utils)
import tqdm
import numpy as np
```

# Read all relevant models

```python
ich360=cobra.io.read_sbml_model('../../Model/iCH360/Escherichia_coli_iCH360.xml')
iml1515=cobra.io.read_sbml_model('../data/metabolic_models/iML1515/iML1515.xml')
```

```python

ich360.reactions.EX_glc__D_e.lower_bound=0.
iml1515.reactions.EX_glc__D_e.lower_bound=0.
```

## Graphic parameters

```python
aerobic_color="#ff5460ff"
anaerobic_color="#1484ffff"
iml1515_color='#80511f'
ich360red_color=iml1515_color
ecc_color=iml1515_color
ecc2_color='#e92ea3ff'
grid_alpha=0.3
title_fontsize=16
```

## Define key metabolic precursors of interest
We focus on 13 key procursors relevant to a variety of biotechnological applications

```python
precursors = ['g6p_c', 'f6p_c', 'e4p_c', 'r5p_c', 'g3p_c', '3pg_c', 'pep_c', 'pyr_c', 'accoa_c', 'akg_c', 'succoa_c', 'oaa_c', 'cit_c']
```

## Add sinks reactions for each of these

```python

for m in [ich360, iml1515]:
    #add a sink for each target precursor to test
    for p_id in precursors:
        m.add_boundary(metabolite=m.metabolites.get_by_id(p_id),type='sink',lb=0)
    # We also need to add a production source for CoA, since our model cannot produce it de novo, 
    # and in general we are not interested in the CoA prosthetic group for this analysis
    m.add_boundary(metabolite=m.metabolites.get_by_id('coa_c'),type='sink',lb=-1000,ub=0)

products=[f'SK_{p_id}' for p_id in precursors] #names of the traget reactions to use as objective
product_names=[ich360.metabolites.get_by_id(x).name for x in precursors] #name of the products, for titles in the plots
```

## Draw production envelopes

```python
carbon_source='glc__D'
fig,ax=plt.subplots(int(np.ceil(len(products)/3)),3,figsize=(15,10),sharex=True,sharey=True)
plt.tight_layout()

for k,product in tqdm.tqdm(enumerate(products)):
    coords=np.unravel_index(k,ax.shape)
        #iML1515======
    #aerobic
    fba_utils.plot_envelope(wt_model=iml1515,
                        knockouts=[],
                        carbon_sources=[carbon_source],
                        target_reaction=product,
                        ax=ax[coords],
                        color=iml1515_color,
                        ls='--',
                        label='iML1515',
                        fill=False)
    #anaerobic
    fba_utils.plot_envelope(wt_model=iml1515,
                        knockouts=['EX_o2_e'],
                        carbon_sources=[carbon_source],
                        target_reaction=product,
                        ax=ax[coords],
                        color=iml1515_color,
                        label=None,
                        ls='--',
                        fill=False)
    #=======
    #iCH360 (left)======
    #aerobic
    fba_utils.plot_envelope(wt_model=ich360,
                        knockouts=[],
                        carbon_sources=[carbon_source],
                        target_reaction=product,
                        ax=ax[coords],
                        label='aerobic',
                        color=aerobic_color)
    #anaerobic
    fba_utils.plot_envelope(wt_model=ich360,
                        knockouts=['EX_o2_e'],
                        carbon_sources=[carbon_source],
                        target_reaction=product,
                        ax=ax[coords],
                        label='anaerobic',
                        color=anaerobic_color)
    ax[coords].set_title(f'{product_names[k]}',fontsize=16)
    ax[coords].grid(alpha=grid_alpha)

ax[0,0].legend(reverse=True)
fig.suptitle("Growth on glucose",fontsize=title_fontsize,y=1.1)
fig.supxlabel('Growth rate [1/h]',y=-0.02)
fig.supylabel('Production flux [mmol/gDW/h]',x=-0.02)
```

The main differences between the two models can be identified for acetyl-CoA production and anaerobic pyruvate production. These have been already addressed and discussed in the manuscript (see also [acetate_production_iCH360_vs_iML1515](acetate_production_iCH360_vs_iML1515.ipynb))
