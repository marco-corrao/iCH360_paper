# Flux force efficacy ($\eta$) as a function of (scaled) driving forces

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#allow latex in figure labels
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "DeJavu Sans",
})
```

$$
    \eta=\frac{\mathrm{e}^{-\frac{\Delta_rG'}{\rm RT}}-1}
              {\mathrm{e}^{-\frac{\Delta_rG'}{\rm RT}}+1}
        =\tanh\Big(-\frac{1}{2}\frac{\Delta_rG'}{\rm RT}\Big)

$$

```python
x_ar=np.linspace(0,6,100)
y_ar=np.tanh(0.5*x_ar)
fig,ax=plt.subplots(figsize=(6,3))

#ax.hlines(y=[1.0],xmin=0.0,xmax=6,linestyles='dashed',colors='black')
#ax.vlines(x=2*np.arctanh(0.5),ymin=0,ymax=0.5,linestyles='dashed',colors='black')
#ax.hlines(y=[0.5],xmin=0.0,xmax=2*np.arctanh(0.5),linestyles='dashed',colors='black')

ax.plot(x_ar,y_ar,linewidth=3,color='red')
ax.set_xlim(-0.0,6)
ax.set_ylim(-0.0,1.1)

ax.set_yticks([0.0,0.5,1.0])
ax.set_xlabel(r"Driving force ($-\frac{\Delta_rG'}{\rm RT}$)",fontsize=16)
ax.set_ylabel(r"Flux force efficacy ($\eta$)",fontsize=16)
ax.grid()
```
